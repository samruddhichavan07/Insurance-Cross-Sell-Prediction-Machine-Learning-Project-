```python
X = data_modeling[[
    'Gender_Male',
    'Gender_Female',
    'Age_log',
    'Driving_License',
    'Region_Code_Encoding',
    'Previously_Insured',
    'Vehicle_Age_Encoding',
    'Vehicle_Damage_Encoding',
    'Annual_Premium',
    'Policy_Sales_Channel_Encoding',
    'Vintage'
]]

y = data_modeling['Response']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

# Standardization Scaling

```python
FEATURES = [
    'Gender_Male',
    'Gender_Female',
    'Age_log',
    'Driving_License',
    'Region_Code_Encoding',
    'Previously_Insured',
    'Vehicle_Age_Encoding',
    'Vehicle_Damage_Encoding',
    'Annual_Premium',
    'Policy_Sales_Channel_Encoding',
    'Vintage'
]
```

```python
from sklearn.preprocessing import StandardScaler

FEATURES = [
    'Gender_Male',
    'Gender_Female',
    'Age_log',
    'Driving_License',
    'Region_Code_Encoding',
    'Previously_Insured',
    'Vehicle_Age_Encoding',
    'Vehicle_Damage_Encoding',
    'Annual_Premium',
    'Policy_Sales_Channel_Encoding',
    'Vintage'
]

scaler = StandardScaler().fit(X_train[FEATURES])
```

```python
std_means = scaler.mean_
print('Standardization Mean:', std_means)

std_vars = scaler.var_
print('Standardization Var:', std_vars)
```

### Dataset Train

```python
X_train_scaled = scaler.transform(X_train[FEATURES].values)
```

### Dataset Test

```python
X_test_scaled = scaler.transform(X_test[FEATURES].values)
```

# Stage 3

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

### Fit Training Data

```python
rf = RandomForestClassifier(random_state = 42)
rf = rf.fit(X_train_scaled, y_train) 
rf 
```

### Predict Testing Data

```python
test_predicted_rf = rf.predict(X_test_scaled)
```

### Model Evaluation

#### Test

```python
print('\033[1mRandom Forest\033[0;0m')

print('\nConfusion Matrix: Testing Data\n', confusion_matrix(y_test, test_predicted_rf))

print('\nROC AUC Score: Testing Data\n', roc_auc_score(y_test, test_predicted_rf))

print('\nAccuracy: Testing Data\n', accuracy_score(y_test, test_predicted_rf))

print('\nClassification Report: Testing Data\n', classification_report(y_test, test_predicted_rf, digits = 6))
```

#### Train

```python
train_predicted_rf = rf.predict(X_train_scaled)
```

```python
print('\033[1mRandom Forest\033[0;0m')

print('\nConfusion Matrix: Training Data\n', confusion_matrix(y_train, train_predicted_rf))

print('\nROC AUC Score: Training Data\n', roc_auc_score(y_train, train_predicted_rf))

print('\nAccuracy: Training Data\n', accuracy_score(y_train, train_predicted_rf))

print('\nClassification Report: Training Data\n', classification_report(y_train, train_predicted_rf, digits = 6))
```

### Feature Importance

```python
feature_importances = pd.Series(rf.feature_importances_, index = X.columns)

plt.figure(figsize = (20.7, 12))
ax = feature_importances.nlargest(11).plot(kind = 'barh')
ax.invert_yaxis()

plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.xlabel('score', fontsize = 14)
plt.ylabel('feature', fontsize = 14)
plt.title('Random Forest\nFeature Importance Score', fontsize = 24, fontweight = 'bold')

plt.tight_layout()
```

```python
import pandas as pd
import joblib  # for loading your saved model
import os

# -----------------------------
# Path to your model
# -----------------------------
model_path = "backend/artifacts/model_xgb.joblib"

# Optional: check if the file exists
if not os.path.exists(model_path):
    raise FileNotFoundError(f"Model file not found at {model_path}")

# -----------------------------
# Load your trained model
# -----------------------------
model = joblib.load(model_path)

# -----------------------------
# Prepare your new input for YES prediction
# -----------------------------
X_new = pd.DataFrame([{
    'Gender_Male': 1,               # male customer
    'Gender_Female': 0,
    'Age_log': 4.0,                  # log(age) ~55 years old
    'Driving_License': 1,
    'Region_Code_Encoding': 28,      # high-frequency region
    'Previously_Insured': 0,         # new customer → more likely YES
    'Vehicle_Age_Encoding': 1,       # newer car
    'Vehicle_Damage_Encoding': 1,    # car was damaged → critical for YES
    'Annual_Premium': 40000,         # typical premium
    'Policy_Sales_Channel_Encoding': 152, # popular sales channel
    'Vintage': 30                     # loyalty factor
}])

# -----------------------------
# If a scaler was used, scale the input
# -----------------------------
scaler_path = "backend/artifacts/scaler.joblib"
scaler = joblib.load(scaler_path) if os.path.exists(scaler_path) else None
X_new_scaled = scaler.transform(X_new) if scaler else X_new.values

# -----------------------------
# Predict probability
# -----------------------------
prob = model.predict_proba(X_new_scaled)[:, 1]
print("Probability of Interested:", round(prob[0]*100, 2), "%")

# -----------------------------
# Predict class with threshold 0.5
# -----------------------------
threshold = 0.01
prediction = (prob >= threshold).astype(int)
status = "YES" if prediction[0] == 1 else "NO"
print("Predicted Class:", status)
```
